```python
#sol1 and 2) 

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def search_amazon(product_name):
    # Set up Selenium WebDriver
    driver=webdriver.Chrome()
    driver.maximize_window()
    driver.get("https://www.amazon.in/")
    
    try:
        # Find the search bar and enter the product name
        search_box = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "twotabsearchtextbox")))
        search_box.send_keys(product_name)
        search_box.send_keys(Keys.RETURN)

        # Wait for the search results to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div[data-component-type='s-search-result']")))
        
        # Find and print product details
        products = driver.find_elements(By.CSS_SELECTOR, "div[data-component-type='s-search-result']")
        if products:
            for product in products:
                product_brand = product.find_element(By.CSS_SELECTOR, "span.a-size-base-plus.a-color-base.a-text-normal").text.strip()
                product_title = product.find_element(By.CSS_SELECTOR, "span.a-text-normal").text.strip()
                product_link = product.find_element(By.CSS_SELECTOR, "a.a-link-normal").get_attribute("href")
                product_price = product.find_element(By.CSS_SELECTOR, "span.a-price span.a-offscreen").text.strip()
                return_exchange = product.find_element(By.CSS_SELECTOR, "div.a-row.a-size-base.a-color-secondary").text.strip()
                expected_delivery = product.find_element(By.CSS_SELECTOR, "span.a-text-bold").text.strip()
                # availability = product.find_element(By.CSS_SELECTOR, "span.a-size-small.a-color-secondary").text.strip()
                print(f"Brand Name: {product_brand}\nProduct: {product_title}\nPrice: {product_price}\nLink: {product_link}\nReturn Exchange: {return_exchange}\nExpected Delivery: {expected_delivery}\n")
        else:
            print("No products found.")
    
    finally:
        # Close the browser
        driver.quit()

if __name__ == "__main__":
    user_input = input("Enter the product name to search on Amazon.in: ")
    search_amazon(user_input)
```

```python
#sol3) 

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time
import os
import requests

# Function to create directory if it doesn't exist
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Function to download images
def download_images(url, directory, filename):
    response = requests.get(url)
    with open(os.path.join(directory, filename), 'wb') as file:
        file.write(response.content)

# Function to scrape images
def scrape_images(keyword, num_images):
    # Start a Chrome WebDriver
    driver = webdriver.Chrome()
    driver.maximize_window()

    # Access Google Images
    driver.get("https://images.google.com/")
    time.sleep(2)  # Let the page load

    # Find the search bar and enter the keyword
    search_bar = driver.find_element(By.ID, "APjFqb")
    search_bar.send_keys(keyword + " images")
    search_bar.send_keys(Keys.RETURN)
    time.sleep(2)  # Let the page load

  
    # Scroll to load more images
    for _ in range(3):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)

    # Find all the image elements
    images = driver.find_elements(By.CSS_SELECTOR, "Q4gaf")

    # Create a directory for the keyword
    directory = f"{keyword}_images"
    create_directory(directory)

    # Download the images
    for i, image in enumerate(images[:num_images]):
        image_url = image.get_attribute('src')
        if image_url:
            download_images(image_url, directory, f"{keyword}_{i}.jpg")

    # Close the WebDriver
    driver.quit()

# Main function
if __name__ == "__main__":
    keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
    num_images_per_keyword = 10

    for keyword in keywords:
        scrape_images(keyword, num_images_per_keyword)
```

```python
def get_coordinates(city):
    driver = webdriver.Chrome()
    driver.get("https://www.google.com/maps")
    
    search_box = driver.find_element(By.XPATH,'//*[@id="searchboxinput"]')
    search_box.send_keys(city)
    search_box.send_keys(Keys.RETURN)
    
    current_url = driver.current_url  # Find the current URL which contains the coordinates
    lat_lng_index = current_url.find('@') + 1
    lat_lng_str = current_url[lat_lng_index:].split(',')[0:2]
    latitude = lat_lng_str[0]
    longitude = lat_lng_str[1]
    
    
    return latitude, longitude 

    
    
```

```python
city = input("Enter the name of the city: ")
latitude, longitude = get_coordinates(city)
print("Latitude:", latitude)
print("Longitude:", longitude)
```

```python
# Set up the WebDriver
driver = webdriver.Chrome()  # Replace with the path to your WebDriver executable
```

```python
# Open the YouTube video
video_url = 'https://www.youtube.com/watch?v=wa--U0DVyRU'  # Replace with the URL of the YouTube video
driver.get(video_url)
```

```python
# Scroll to load comments
scroll_pause_time = 2  # Adjust the pause time as needed
scrolls = 10  # Adjust the number of scrolls as needed

for _ in range(scrolls):
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(scroll_pause_time)
```

```python
# Extract comments, upvotes, and time
comments = driver.find_elements(By.XPATH, '//div[@class="style-scope ytd-expander"]')
upvotes = driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-engagement-bar"]')
times = driver.find_elements(By.ID,'published-time-text')
```

```python
# Store the extracted data
extracted_data = []
for comment, upvote, time in zip(comments, upvotes, times):
    extracted_data.append({
    'comment': comment.text,
    'upvote': upvote.text,
    'time': time.text
    })
```

```python
# Close the WebDriver
driver.quit()
```

```python
# Print the extracted data
for data in extracted_data[:500]:
    print(data)
```

```python
#sol7)

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

def scrape_forbes_billionaires():
    options = Options()
    options.headless = True

    driver = webdriver.Chrome(options=options)

    url = "https://www.forbes.com/billionaires/"
    driver.get(url)
    time.sleep(5) # Let the page load

    billionaires = []

    rows = driver.find_elements(By.XPATH, "//div[@class='Table_tableRow__lF_cY']")
    print("Number of rows found:", len(rows))
    for row in rows:
        rank = row.find_element(By.XPATH, ".//div[@class='Table_rank__X4MKf']")
        name = row.find_element(By.XPATH, ".//div[@class='Table_personName__Bus2E']")
        net_worth = row.find_element(By.XPATH, ".//div[@class='Table_finalWorth__UZA6k']")
        
        
        billionaire = {
            "Rank": rank.text,
            "Name": name.text,
            "Net worth": net_worth.text,
        
        }
        billionaires.append(billionaire)

    driver.quit()
    return billionaires

billionaires_data = scrape_forbes_billionaires()
if billionaires_data:
    for billionaire in billionaires_data:
        print(billionaire)
else:
    print("No billionaires data found.")


```

```python
#sol6)

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

def scrape_best_gaming_laptops():
    options = Options()
    options.headless = True

    driver = webdriver.Chrome(options=options)

    url = "https://www.digit.in/top-products/best-gaming-laptops-40.html"
    driver.get(url)
    time.sleep(5) 

    laptops = []

    laptop_elements = driver.find_elements(By.XPATH, "//div[@class='rh_grid_image_3_col']")
    print("Number of laptop elements found:", len(laptop_elements))
    for laptop_element in laptop_elements:
        name = laptop_element.find_element(By.XPATH, ".//div[@class='rh_gr_top_middle mb10 colored_rate_bar']//a").text
        display_size = laptop_element.find_element(By.XPATH, ".//span[@class='meta_v_value']").text
        operating_system = laptop_element.find_element(By.XPATH, ".//span[@class='meta_v_value']").text
        processor = laptop_element.find_element(By.XPATH, ".//div[@class='woo_code_loop_item font90']").text

        laptop = {
            "Name": name,
            "Processor": processor,
            "Display Size": display_size,
            "Operating System": operating_system,
        }
        laptops.append(laptop)

        driver.quit()
        return laptops

best_gaming_laptops = scrape_best_gaming_laptops()
if best_gaming_laptops:
    for laptop in best_gaming_laptops:
        print(laptop)
else:
    print("No laptops data found.")
```
