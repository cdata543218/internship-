```python
#sol1)


from selenium import webdriver
from selenium.webdriver.common.by import By


driver = webdriver.Chrome()  


url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"


driver.get(url)


table_rows = driver.find_elements(By.CSS_SELECTOR, ".wikitable tbody tr")

for row in table_rows[1:]:

    rank = row.find_element(By.CSS_SELECTOR, "td:nth-of-type(1)").text.strip()
    name = row.find_element(By.CSS_SELECTOR, "td:nth-of-type(2)").text.strip()
    artist = row.find_element(By.CSS_SELECTOR, "td:nth-of-type(3)").text.strip()
    upload_date = row.find_element(By.CSS_SELECTOR, "td:nth-of-type(4)").text.strip()
    views = row.find_element(By.CSS_SELECTOR, "td:nth-of-type(5)").text.strip()

    print("Rank:", rank)
    print("Name:", name)
    print("Artist:", artist)
    print("Upload Date:", upload_date)
    print("Views:", views)
    print()


driver.quit()
```

```python
table_rows=driver.find_elements(By.XPATH,"//table[@id='table_id']/tbody/tr")
```

```python
rank = []
state = []
gsdp_18_19 = []
gsdp_19_20 = []
share_18_19 = []
gdp_billion = []
```

```python
for row in table_rows:
    columns = row.find_elements(By.TAG_NAME, "td")
    rank.append(columns[0].text)
    state.append(columns[1].text)
    gsdp_18_19.append(columns[3].text)
    gsdp_19_20.append(columns[3].text)
    share_18_19.append(columns[1].text)
    gdp_billion.append(columns[1].text)
```

```python
#sol4


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time

# Set up the Chrome webdriver
chrome_options = Options()
chrome_options.add_argument("--headless")  
service = Service()  
driver = webdriver.Chrome(service=service, options=chrome_options)

url = "https://github.com/trending"
driver.get(url)

time.sleep(5)


trending_repositories = driver.find_elements(By.CSS_SELECTOR, 'article.Box-row')

if trending_repositories:
    
    for repository in trending_repositories:
        
        title_element = repository.find_element(By.CSS_SELECTOR, 'h2 a')
        title = title_element.text.strip()

        
        description_element = repository.find_element(By.CSS_SELECTOR, 'p')
        description = description_element.text.strip()

        
        contributors_element = repository.find_element(By.CSS_SELECTOR, 'a.Link--muted')
        contributors_count = contributors_element.text.strip()

        
        language_element = repository.find_element(By.CSS_SELECTOR, "span[itemprop='programmingLanguage']")
        language = language_element.text.strip() 

        
        print("Repository title:", title)
        print("Description:", description)
        print("Contributors count:", contributors_count)
        print("Language use:", language)
        print()
else:
    print("No trending repositories found.")


driver.quit()
```

```python
#sol6)

from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare"
driver.get(url)

table = driver.find_element(By.XPATH, "//table[@class='in-article sortable']")

rows = table.find_elements(By.XPATH, ".//tr[position() > 1]")

for row in rows:
    cells = row.find_elements(By.XPATH, ".//td")
    
    book_name = cells[1].text
    author_name = cells[2].text
    volumes_sold = cells[3].text
    publisher = cells[4].text
    genre = cells[5].text
    
    print("Book Name:", book_name)
    print("Author Name:", author_name)
    print("Volumes Sold:", volumes_sold)
    print("Publisher:", publisher)
    print("Genre:", genre)
    print()

driver.quit()
```

```python
#sol7) 

from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

url = "https://www.imdb.com/list/ls512407256/"
driver.get(url)

tv_series_items = driver.find_elements(By.CLASS_NAME, "lister-item-content")

for item in tv_series_items:
    name = item.find_element(By.TAG_NAME, "h3").text.strip()

    year_span = item.find_element(By.CLASS_NAME, "lister-item-year").text.strip()

    genre = item.find_element(By.CLASS_NAME, "genre").text.strip()

    runtime = item.find_element(By.CLASS_NAME, "runtime").text.strip()

    ratings = item.find_element(By.CLASS_NAME, "ipl-rating-star__rating").text.strip()
    
    votes = item.find_element(By.CSS_SELECTOR, "span[name='nv']").text.strip()

    print("Name:", name)
    print("Year Span:", year_span)
    print("Genre:", genre)
    print("Run Time:", runtime)
    print("Ratings:", ratings)
    print("Votes:", votes)
    print()

driver.quit()
```
